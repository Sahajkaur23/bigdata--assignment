```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x1 = np.array([5.3, 4.8, 2, 3, 4])
x2 = np.array([9.5, 11, 14, 8, 12])
y = np.array([127, 160, 150, 170, 140])

x1_mean = np.mean(x1)
x2_mean = np.mean(x2)
y_mean  = np.mean(y)

numerator=0
denominator=0

for i in range(len(x1)):
    numerator += (x1[i] - x1_mean) * (y[i] - y_mean)
    denominator += (x1[i] - x1_mean) ** 2

b1 = numerator/denominator

b2 = np.sum((x2 - np.mean(x2)) * (y - np.mean(y))) / np.sum((x2 - np.mean(x2)) ** 2)
b0=y_mean-(b1*x1_mean)-(b2*x2_mean)
Y=b0+b1*x1*b2*x2
print("value of b1 :",b1)
print('value of b2 :',b2)
print('value of b0 :',b0)


plt.scatter(x1, y, color='blue', label='Data Points')
plt.plot(x1, Y, color='red', label='Multiple Linear Regression Line')

plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Multiple Linear Regression')


plt.grid(True)
plt.show()
```
